```python
!pip install vaderSentiment
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer as VADER
import matplotlib.pyplot as plt

texts = [
    "I love this brand!",
    "This product is terrible",
    "I'm not sure about this",
    "Absolutely fantastic experience",
    "Worst service ever"
]

df = pd.DataFrame({'text': texts})

analyzer = VADER()
df['score'] = df['text'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
df['sentiment'] = df['score'].apply(lambda x: 'positive' if x > 0.05 else ('negative' if x < -0.05 else 'neutral'))

counts = df['sentiment'].value_counts()
plt.bar(counts.index, counts.values, color=['green', 'red', 'gray'])
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Sentiment Distribution')
plt.show()
```
